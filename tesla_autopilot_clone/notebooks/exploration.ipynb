```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import Image, display

# Load the model (downloads weights automatically on first run)
model = YOLO('yolov8n.pt')
print('Model loaded:', model.info())
```

```python
# 2. Run inference on a sample image
IMAGE_PATH = 'assets/sample.jpg'   # replace with your image
results = model(IMAGE_PATH, conf=0.40)[0]

# Display annotated result
annotated = results.plot()
plt.figure(figsize=(14, 8))
plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('YOLOv8 Detections')
plt.show()
```

```python
# 3. Inspect detection data
for box in results.boxes:
    cls   = int(box.cls[0])
    conf  = float(box.conf[0])
    xyxy  = box.xyxy[0].tolist()
    label = model.names[cls]
    print(f'{label:15s}  conf={conf:.3f}  box={[round(v,1) for v in xyxy]}')
```

```python
# 4. Confidence distribution plot
confs = [float(b.conf[0]) for b in results.boxes]
plt.figure(figsize=(8, 4))
plt.hist(confs, bins=20, color='steelblue', edgecolor='white')
plt.xlabel('Confidence'); plt.ylabel('Count')
plt.title('Confidence Score Distribution')
plt.tight_layout(); plt.show()
```

```python
# 5. Class frequency bar chart
from collections import Counter
class_counts = Counter(model.names[int(b.cls[0])] for b in results.boxes)

plt.figure(figsize=(10, 4))
plt.bar(class_counts.keys(), class_counts.values(), color='coral', edgecolor='white')
plt.xlabel('Class'); plt.ylabel('Count')
plt.title('Detected Classes')
plt.xticks(rotation=30, ha='right')
plt.tight_layout(); plt.show()
```
